```python
df_sum_null = df.isnull().sum()
df_sum_null
```

```python
df_sum_null[0:15]
```

```python
#formula for finding missing no: of cell's percentage
# (total missing values/total no of cells)*100

#df_sum_null = df.isnull().sum()
total_missing_values = df_sum_null.sum()

total_no_of_cells = np.product(df.shape)

percentage_missing = (total_missing_values/total_no_of_cells)*100

percentage_missing
```

```python
# if am Droping missing values: rows
#ie, drop the rows with missing value, know what happen
dropna_byrows = df.dropna()
dropna_byrows
```

```python
# if am Droping missing values: Column
#ie, drop the column with missing value, know what happen
dropna_by_column = df.dropna(axis=1)

#NOW checking how many col we lost
original_no_of_column = df.shape[1]
dropna_no_of_column = dropna_by_column.shape[1]
#The first element (df.shape[0]) is the number of rows.
#The second element (df.shape[1]) is the number of columns.

no_of_col = original_no_of_column - dropna_no_of_column
```

```python
no_of_col
#we have 65 col after dropping the dropna
```

```python
#Fill in missing values automatically and then replacing any remainig nan with zero = imputation

df_new = df.fillna(method='bfill',axis=0).fillna(0)
df_new
```
